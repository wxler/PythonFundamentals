# 2. 条件判断 `if`

`if` 的语法
``` python
if 条件 :
    # some code
elif 条件二： #elif 不用加空格
    # some code
else:
    # other code
```
 
注意点：冒号 缩进 执行顺序

````python
num1, num2 = 3 , 5  #两个变量一起赋值的语句
if num1 > num2:
    print('第一个数是大于第二个数的')
else:
    print('第一个数小于或等于第二个数')
````

例：判断是否是闰年

````python
year = int(input("输入一个年份: ")) 
if (year % 4) == 0:
    if (year % 100) == 0:
        if (year % 400) == 0:
            print("{0} 是闰年".format(year))   # 整百年能被400整除的是闰年
        else:
            print("{0} 不是闰年".format(year))
    else:
        print("{0} 是闰年".format(year))       # 非整百年能被4整除的为闰年
else:
    print("{0} 不是闰年".format(year))
````

````python
# 简化版的代码
year = int(input("请输入一个年份："))
if (year % 4) == 0 and (year % 100) != 0 or (year % 400) == 0:   #外面可以不要大括号
    print("{0}是闰年".format(year))  
    #0代表位置，不要也可以，默认第一个{}是第一个位置，默认第二个{}是第二个位置
else:
    print("{0}不是闰年".format(year))
````

````python
def isleap(year):
    if (year % 4) == 0 and (year % 100) != 0 or (year % 400) == 0:
        print("{0}是闰年".format(year))
    else:
        print("{0}不是闰年".format(year))
````

### 5.3.5 不定长参数


- 定义不定长参数时，该参数前加\*或\*\*，当传入实参多于形参时，超出部分记入不定长参数
- 形参前面加*，接收的参数转为元组
- 形参前面加**，接收的参数转化为字典,但参数须以赋值形式传入

````python
def my_print2(x, *y ,**z):
    print('x:', x, type(x)) #print可以无限逗号隔开输出
    print('y:', y, type(y))
    print('z:', z, type(z))
````

````python
my_print2(1,2,3,'sd',55,'sdfdsa')
````

````python
my_print2(1,2,3,'ttt','sdaf',name = 'jack', year = 15)
````

我还想看打架

````python
class Hero(People):
    def __init__(self,name, hp, damage, country):
        People.__init__(self, name, hp, damage)
        self.country = country
        
    def get_inf(self):  #不要忘记self
        """获取实例信息"""
        print("姓名：{}".format(self.name))
        print("攻击力：{}".format(self.damage))
        print("当前血量：{}".format(self.hp))
        print("阵营：{}".format(self.country))   
        
    def attack(self,enemy):
        """self攻击enemy
           enemy血量减少
           enemy血量计算方式：enemy当前血量 - self攻击力
           """
        print(self.name)
        print("攻击力为：{}".format(self.damage))
        print("{}目前血量为：{}".format(enemy.name, enemy.hp))
        print("{}攻击{}".format(self.name,enemy.name))
        enemy.hp-=self.damage
        print("{}剩余血量为：{}".format(enemy.name, enemy.hp))
        print("***********************************")
````

````python
#创建实例
liufeng = Hero('刘封', 4, 2, '蜀')
ganning = Hero('甘宁', 4, 2, '吴')
````

````python
#产看ganning被打前信息
ganning.get_inf()
print('#################')
#liufeng攻击ganning
liufeng.attack(ganning)
#查看ganning被打后信息
ganning.get_inf()
````
